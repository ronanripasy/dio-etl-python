## Limpando dados NULL

```python
# Atribuindo <NA> para valores não informados
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
```

```python
# Atribuindo <NA> para valores faltantes
df.replace(['**', '***', '###!', '####', '****', '*****', 'NULL'], pd.NA, inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# alterando valores <NA> para zero
df.fillna(0, inplace=True)

# Alterando valores <NA> para 10 na coluna: total_recomendacoes
df.fillna(value={'total_recomendacoes': 10}, inplace=True)
```

```python
# Criando nova coluna no DataFrame
df['total_recomendacoes_bk'] = df.total_recomendacoes
```

```python
df
```

```python
# Apagando a coluna informada 
df.drop(['total_recomendacoes_bk'], axis=1, inplace=True)
```

```python
# Apagando todas as linhas com valor <NA> / NULL
df.dropna()
```
